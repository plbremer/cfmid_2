```python
import pandas as pd
```

```python
from rdkit import Chem
from rdkit.Chem import Draw
import matplotlib.pyplot as plt
```

```python
input_file_address='./input.txt'
```

```python
bit_panda=pd.read_csv(input_file_address,sep='\t')
```

```python
bit_panda
```

```python
for index, series in bit_panda.iterrows():
    if index==0 or index==9:
        continue
    else:
        temp_mol=Chem.MolFromSmiles(series['bit_description'])
        #Draw.MolToImage(temp_mol)
        Draw.MolToFile(temp_mol,'./'+str(series['associated_with'])+'_'+str(series['bit_number'])+'.png',size=(100,100))
```
