```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
my_OrdinalEncoder=OrdinalEncoder()
```

```python
classes_as_numbers=my_OrdinalEncoder.fit_transform(combined_panda['Superclass'].to_numpy().reshape(-1,1))
```

```python
combined_panda['superclass_numbers']=classes_as_numbers
```

```python
import matplotlib.pyplot as plt
```

```python
x_direction_bin_count=20
y_direction_bin_count=len(combined_panda.superclass_numbers.unique())+1
```

```python
def make_heatmap(input_panda):
    my_histogram, x_edges, y_edges = np.histogram2d(
        input_panda['dot_product'], input_panda['superclass_numbers'], bins=[x_direction_bin_count, y_direction_bin_count]
    )
    my_histogram=np.divide(my_histogram,np.sum(my_histogram,axis=1).reshape(-1,1))
    my_histogram=np.divide(my_histogram,np.sum(my_histogram,axis=0))
    extent = [x_edges[0], x_edges[-1], y_edges[0], y_edges[-1]]
    # create figure from out histogram
    fig, ax = plt.subplots()
    divider = make_axes_locatable(ax)
    plt.title('title')
    plt.ylabel("...")
    plt.xlabel("dot product")
    tick_spacing = 2
    ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
    cax = divider.append_axes("right", size="5%", pad=0.05)
    image = ax.imshow(
        my_histogram.T, extent=extent, origin="lower", aspect="auto", cmap="magma"#, vmin=0, vmax=1
    )
    fig.colorbar(image, cax=cax, orientation="vertical")    
    plt.show()
```

```python
make_heatmap(combined_panda)
```
