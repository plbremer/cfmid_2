```python
x_direction_bin_count=20
y_direction_bin_count=len(combined_panda.superclass_numbers.unique())+1
```

```python
combined_panda[['Superclass','superclass_numbers']].value_counts()
```

```python
combined_panda.superclass_numbers=combined_panda.superclass_numbers.map(
    {
        10.0:0.0,
        1.0:1.0,
        6.0:2.0,
        13.0:3.0,
        8.0:4.0,
        7.0:5.0,
        3.0:6.0,
        12.0:7.0,
        9.0:8.0,
        5.0:9.0,
        0.0:10.0,
        2.0:11.0,
        11.0:12.0,
        4.0:13.0
    }
)
```

```python
def make_heatmap(input_panda,translation_list):
    my_histogram, x_edges, y_edges = np.histogram2d(
        input_panda['dot_product'], 
        input_panda['superclass_numbers'], 
        bins=[x_direction_bin_count, len(input_panda.superclass_numbers.unique())]
    )
    my_histogram=np.divide(my_histogram,np.sum(my_histogram,axis=1).reshape(-1,1))
    #my_histogram=np.divide(my_histogram,np.sum(my_histogram,axis=0))
    #######
    #transform to percents
    my_histogram=100*my_histogram
    plt.rcParams['font.size']=18
    #######
    extent = [x_edges[0], x_edges[-1], y_edges[0], y_edges[-1]]
    # create figure from out histogram
    fig, ax = plt.subplots()
    divider = make_axes_locatable(ax)
    plt.title('title')
    #plt.ylabel("...")
    plt.xlabel("dot product")
    tick_spacing = 2
    ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
    cax = divider.append_axes("right", size="5%", pad=0.05)
    image = ax.imshow(
        my_histogram.T, extent=extent, origin="lower", aspect="auto", cmap="magma"#, vmin=0, vmax=1
    )
    #######
    for i in range(1,len(x_edges)-1):
        ax.vlines(
            x=x_edges[i],
            ymin=y_edges[0],
            ymax=y_edges[-1],
            colors='w'
        )
    ax.set_xticks([ (x_edges[i]-(x_edges[1]/2)) for i in range(1,len(x_edges))])
    ax.set_xticklabels([ (str(1000-i*50+25)) for i in range(1,len(x_edges))],rotation=-90)  
    
    #ax.set_yticks([0.5+i for i in range(len(input_panda.superclass_numbers.unique()))])
    #ax.set_yticks(np.arange())
    number_of_classes=len(input_panda.superclass_numbers.unique())
    ax.set_yticks(
        np.arange(
            ((number_of_classes-1)/number_of_classes)-0.5*((number_of_classes-1)/number_of_classes),
            number_of_classes-1,
            (number_of_classes-1)/number_of_classes
        )
    )
    ax.set_yticklabels(translation_list)
    
    
    fig.colorbar(image, cax=cax, orientation="vertical",label='Percentage of Compounds')
    #######   
    plt.savefig('../../../../results/final_figures/submitted_to_manuscript/[M+H]+_hcd_clases_vs_dot_product.png')
    plt.savefig('../../../../results/final_figures/submitted_to_manuscript/[M+H]+_hcd_clases_vs_dot_product.eps')
```
