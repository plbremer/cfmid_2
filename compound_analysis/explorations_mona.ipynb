```python
temp['Collision_energy_parsed']=temp['Collision_energy'].str[0:2]
```

```python
inchikeys=temp.InChIKey.unique()
```

```python
results={
    'inchikey':[],
    'dot_product':[],
    'collision_energy_exp':[],
    'predicted_optimal_nce':[]
}
for inchikey in inchikeys:
    #print(temp.loc[temp.InChIKey==inchikey])
    ce_choices=temp.loc[temp.InChIKey==inchikey]['Collision_energy_parsed'].astype(float)
    #print(ce_choices)
    optimal_nce=temp.loc[temp.InChIKey==inchikey]['optimal_nce'].astype(float).to_list()[0]
    #print(optimal_nce)
    nce_error=abs(ce_choices-optimal_nce)
    #print(nce_error)
    #optimal_index=temp.loc[() & ]
    optimal_index=nce_error.loc[nce_error==min(nce_error)].index[0]
    #print(optimal_index)
    #hold=input('hold')
    
    results['inchikey'].append(temp.at[optimal_index,'InChIKey'])
    results['dot_product'].append(temp.at[optimal_index,'dot_product'])
    results['collision_energy_exp'].append(temp.at[optimal_index,'Collision_energy_parsed'])
    results['predicted_optimal_nce'].append(temp.at[optimal_index,'optimal_nce'])
```

```python
result_panda=pd.DataFrame.from_dict(results)
```
